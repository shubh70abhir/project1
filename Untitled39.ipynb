```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to [0, 1]
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Reshape data for Neural Network (NN) and CNN
# For NN, add a channel dimension (not necessary but keeps code consistent)
x_train_nn = x_train.reshape(-1, 28, 28)
x_test_nn = x_test.reshape(-1, 28, 28)

# For CNN, add channel dimension (grayscale=1)
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)
```

```python
model_nn = Sequential([
    Flatten(input_shape=(28, 28)),  # Input layer
    Dense(128, activation='relu'),  # Hidden layer 1
    Dense(64, activation='relu'),   # Hidden layer 2
    Dense(10, activation='softmax') # Output layer
])

model_nn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),  # Conv layer 1
    Conv2D(64, (3, 3), activation='relu'),                          # Conv layer 2
    MaxPooling2D((2, 2)),                                           # Pooling layer
    Dropout(0.25),                                                  # Regularization
    Flatten(),                                                      # Flatten for dense layers
    Dense(128, activation='relu'),                                  # Dense layer
    Dropout(0.5),                                                   # Regularization
    Dense(10, activation='softmax')                                 # Output layer
])

model_cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train the Neural Network
history_nn = model_nn.fit(
    x_train_nn, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32
)

# Train the CNN
history_cnn = model_cnn.fit(
    x_train_cnn, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32
)
```

```python
# Evaluate NN
test_loss_nn, test_acc_nn = model_nn.evaluate(x_test_nn, y_test)
print(f'NN Test Accuracy: {test_acc_nn:.4f}')

# Evaluate CNN
test_loss_cnn, test_acc_cnn = model_cnn.evaluate(x_test_cnn, y_test)
print(f'CNN Test Accuracy: {test_acc_cnn:.4f}')
```

```python
# Plot NN training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_nn.history['accuracy'], label='NN Training Accuracy')
plt.plot(history_nn.history['val_accuracy'], label='NN Validation Accuracy')
plt.title('Neural Network Accuracy')
plt.legend()

# Plot CNN training history
plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['accuracy'], label='CNN Training Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='CNN Validation Accuracy')
plt.title('CNN Accuracy')
plt.legend()
plt.show()
```
